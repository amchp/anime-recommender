```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('datos/users-details-2023.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
plt.hist(data['Completed'], bins=500)
plt.xlabel('Mean Score')
plt.ylabel('Frecuency')
plt.title('Mean Score Histogram')
plt.xlim(0, 2000)
plt.show()
```

```python
plt.hist(data['Rewatched'], bins=1600)
plt.xlabel('Rewatched')
plt.ylabel('Frecuency')
plt.title('Rewatched Histogram')
plt.xlim(0, 250)
plt.show()
```
